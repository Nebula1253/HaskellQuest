```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np

from scipy.stats import pearsonr
```

```python
singleplayer_y1 = pd.read_excel('singleplayer_y1.xlsx')
singleplayer_y2 = pd.read_excel('singleplayer_y2.xlsx')
multiplayer_y2 = pd.read_excel('multiplayer_y2.xlsx')
```

```python
prior_haskell_enjoy_sp_y1 = [[0] * 6 for i in range(6)] 

for _, row in singleplayer_y1.iterrows():
    prior_haskell_enjoy_sp_y1[row['prior_haskell'] - 1][row['how_enjoyable'] - 1] += 1

prior_haskell_enjoy_sp_y1 = np.array(prior_haskell_enjoy_sp_y1).T
```

```python
prior_haskell_enjoy_sp_y2 = [[0] * 6 for i in range(6)] 

for _, row in singleplayer_y2.iterrows():
    prior_haskell_enjoy_sp_y2[row['prior_haskell'] - 1][row['how_enjoyable'] - 1] += 1

prior_haskell_enjoy_sp_y2 = np.array(prior_haskell_enjoy_sp_y2).T
```

```python
prior_haskell_enjoy_mp_y2 = [[0] * 6 for i in range(6)] 

for _, row in multiplayer_y2.iterrows():
    prior_haskell_enjoy_mp_y2[row['prior_haskell'] - 1][row['how_enjoyable'] - 1] += 1

prior_haskell_enjoy_mp_y2 = np.array(prior_haskell_enjoy_mp_y2).T
```

```python
fig, (ax, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(24,7))

sns.heatmap(data=prior_haskell_enjoy_sp_y1, cmap="Blues", fmt='s', 
            annot=np.where(prior_haskell_enjoy_sp_y1 == 0, "", prior_haskell_enjoy_sp_y1), 
            cbar=False, ax=ax, xticklabels=range(1, 7), yticklabels=range(1, 7),
            annot_kws={"size": 12})
ax.invert_yaxis()
ax.set_xlabel("Prior Haskell experience", fontsize=14)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=12)
ax.set_ylabel("Enjoyability rating", fontsize=14)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=12)
ax.set_title('Year 1 Single-Player', fontsize=16)

sns.heatmap(data=prior_haskell_enjoy_sp_y2, cmap="Greens", fmt='s', 
            annot=np.where(prior_haskell_enjoy_sp_y2 == 0, "", prior_haskell_enjoy_sp_y2), 
            cbar=False, ax=ax2, vmax=4, xticklabels=range(1, 7), yticklabels=range(1, 7),
            annot_kws={"size": 12})
ax2.invert_yaxis()
ax2.set_xlabel("Prior Haskell experience", fontsize=14)
ax2.set_xticklabels(ax2.get_xticklabels(), fontsize=12)
ax2.set_ylabel("Enjoyability rating", fontsize=14)
ax2.set_yticklabels(ax2.get_yticklabels(), fontsize=12)
ax2.set_title('Year 2 Single-Player', fontsize=16)

sns.heatmap(data=prior_haskell_enjoy_mp_y2, cmap="Reds", fmt='s', 
            annot=np.where(prior_haskell_enjoy_mp_y2 == 0, "", prior_haskell_enjoy_mp_y2), 
            cbar=False, ax=ax3, vmax=4, xticklabels=range(1, 7), yticklabels=range(1, 7),
             annot_kws={"size": 12})
ax3.invert_yaxis()
ax3.set_xlabel("Prior Haskell experience", fontsize=14)
ax3.set_xticklabels(ax3.get_xticklabels(), fontsize=12)
ax3.set_ylabel("Enjoyability rating", fontsize=14)
ax3.set_yticklabels(ax3.get_yticklabels(), fontsize=12)
ax3.set_title('Year 2 Multiplayer', fontsize=16)

# ax.scatter(singleplayer_y2['prior_program'], singleplayer_y2['how_enjoyable'])
# ax.scatter(multiplayer_y2['prior_program'], multiplayer_y2['how_enjoyable'])

plt.savefig('priorhaskell-enjoyability.png', dpi=300, bbox_inches='tight')
```

```python
from scipy.stats import ttest_ind

def helper(x: str):
    if (x != 'progression_difficulty'):
        print(ttest_ind(singleplayer_y1[x], singleplayer_y2[x], alternative='less', equal_var=False))
        print(ttest_ind(singleplayer_y1[x], multiplayer_y2[x], alternative='less', equal_var=False))
    print(ttest_ind(singleplayer_y2[x], multiplayer_y2[x], alternative='less', equal_var=False))
    print()

helper('how_enjoyable')
helper('intro_effective')
helper('challenge_difficulty')
helper('progression_difficulty')
helper('tutorial_text_rating')
```
